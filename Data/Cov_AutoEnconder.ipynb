```python
import tensorflow
import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Model, Sequential
from keras.layers import Dense, Conv2D, Dropout, BatchNormalization, Input, Reshape, Flatten, Deconvolution2D, Conv2DTranspose, MaxPooling2D, UpSampling2D
from keras.layers.advanced_activations import LeakyReLU
from keras.optimizers import adam
```

#### Encoder

```python
inp = Input((100, 100,3))
enco = Conv2D(64, (3, 3), activation='relu')(inp)
enco = MaxPooling2D((2, 2))(enco)
enco = Conv2D(32, (3, 3), activation='relu')(enco)
enco = MaxPooling2D((2, 2))(enco)
enco = Conv2D(16, (3, 3), activation='relu')(enco)
enco = MaxPooling2D((2, 2))(enco)
l = Flatten()(enco)
l = Dense(625, activation='softmax')(l)
```

#### Decoder

```python
deco = Reshape((25,25,1))(l)
deco = Conv2DTranspose(64,(3, 3), strides=2, activation='relu', padding='same')(deco)
deco = BatchNormalization()(deco)
deco = Conv2DTranspose(64,(3, 3), strides=2, activation='relu', padding='same')(deco)
deco = BatchNormalization()(deco)
deco = Conv2DTranspose(32,(3, 3), activation='relu', padding='same')(deco)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(deco)
```

#### Show the shape of the autoencoder.

```python
encoder = Model(inp,l)
ae = Model(inp, decoded)
ae.summary()
```

#### Load the training data.

```python
import pickle

with open('im_dict.pickle','rb') as im_d:
    im_dict = pickle.load(im_d)
    
label_dict = {
                1: 'Surprise',
                2: 'Fear',
                3: 'Disgust',
                4: 'Happiness',
                5: 'Sadness',
                6: 'Anger',
                7: 'Neutral'
              }

Training_list = []

for label in im_dict['train'].keys():
    train_array_path = 'Data_Arrays/train/' + label_dict[label] + '/'
    for im_name in im_dict['train'][label]:
        array_name = im_name.strip('.jpg') + '_aligned.npy'
        im_array = np.load(train_array_path + array_name)
        Training_list.append((im_array,label))
        
train_indices = range(len(Training_list))
np.random.shuffle(train_indices)
Training_features = np.array([Training_list[i][0] for i in train_indices]) 
Training_labels = np.array([Training_list[i][1] for i in train_indices])

print 'Training features shape:',Training_features.shape
print 'Training labels shape:',Training_labels.shape
```

#### Loading the test data.

```python
Test_list = []

for label in im_dict['test'].keys():
    test_array_path = 'Data_Arrays/test/' + label_dict[label] + '/'
    for im_name in im_dict['test'][label]:
        array_name = im_name.strip('.jpg') + '_aligned.npy'
        im_array = np.load(test_array_path + array_name)
        Test_list.append((im_array,label))
        
test_indices = range(len(Test_list))
np.random.shuffle(test_indices)
Test_features = np.array([Test_list[i][0] for i in test_indices]) 
Test_labels = np.array([Test_list[i][1] for i in test_indices])

print 'Test features shape:',Test_features.shape
print 'Test labels shape:',Test_labels.shape
```

#### Normalize the data.

```python
Training_features,Test_features = Training_features/255.0 - 0.5, Test_features/255.0 - 0.5
```

#### Train the auto-encoder.

```python
import os

ae.compile(optimizer='adam',loss='mse')
ae_train = ae.fit(Training_features,
                  Training_features,
                  epochs=10,
                  batch_size=256,
                  shuffle=True,
                  validation_data=(Test_features,Test_features)
                 )
```

#### Record the model weight.

```python
Model_path = 'Auto_Encoder_model/'
if not os.path.exists(Model_path):
    os.system('mkdir ' + Model_path)
    
model_json = ae.to_json()
with open(Model_path + 'model_ae_tex.json','w') as json_file:
    json_file.write(model_json)

ae.save_weights(Model_path + 'model_ae_tex.h5')
```

#### Plot the training loss and test loss.

```python
import matplotlib.pyplot as plt

training_loss = ae_train.history['loss']
test_loss = ae_train.history['val_loss']

plt.figure()
plt.plot(training_loss,label='Training Loss')
plt.plot(test_loss,label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### Apply PCA to analyze the representation and continue reduce the dimension.

```python
Train_matrix = np.array([encoder.predict(np.array([i]))[0] for i in Training_features])
```

```python
eng_values,eng_vectors = np.linalg.eig(np.dot(Train_matrix.T,Train_matrix))
```

```python
eng_tuple = sorted([(eng_values[i],i) for i in range(len(eng_values))],reverse = True)
eng_sum_list = []
eng_sum = 0
for i in range(len(eng_values)):
    eng_sum += eng_tuple[i][0]
    eng_sum_list.append(eng_sum)
    
plt.figure()
plt.plot(range(1,len(eng_values) + 1),eng_sum_list)
plt.xlabel('Principle Componenets Amount')
plt.ylabel('Various')
```

```python
selected_eigen_index = np.array([eng_tuple[i][1] for i in range(13)])
selected_matrix = eng_vectors[selected_eigen_index]
```

#### Get the representation vector of each label.

```python
Representation_path = 'Expression_Representation/'
if not os.path.exists(Representation_path):
    os.system('mkdir ' + Representation_path)

l_index = 0
label_rep_dict = dict()
for label in label_dict.keys():
    expression = label_dict[label]
    array = np.load('Data_Means/train/' + expression + '.npy')
    array = array/255.0 - 0.5
    repre = encoder.predict(np.array([array]))[0]
    repre_PCA = np.dot(selected_matrix,repre)
    one_hot = np.zeros(CLASS_NUM)
    one_hot[l_index] = 1
    repre_PCA = np.array(list(repre_PCA) + list(one_hot))
    label_rep_dict[label] = repre_PCA
    np.save(Representation_path + expression + '.npy',repre_PCA)
    l_index += 1
    
with open('label_representation.pickle','wb') as lr_dic_file:
    pickle.dump(label_rep_dict,lr_dic_file)
```
